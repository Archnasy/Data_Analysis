## Merge function

```python
# Combining a company_X_Order_Report with Courier_Company_invoice by using merge function 
# For merging a data table, i have used a "Order_ID" feature and User-defined variable "o"
O=pd.merge(Orders,invoices,on="Order ID",how="right")

# To combine a user_defined variable "o" of a new data table with Company_X_SKU_Master table by using merge fuction
#For merging a data table, i have used a "SKU" feature and User-defined variable "Ex"
Ex=pd.merge(O,sku_Master,on="SKU",how="left")

# To combine a user_defined variable "Ex" of a new data table with Company_X_Pincode_Zone table by using merge fuction
#For merging a data table, i have used a Customer_Pincode feature and User-defined variable "Ones"
Merged_dataset=pd.merge(Ex,X_post,how="inner",on="Customer Pincode")
```
